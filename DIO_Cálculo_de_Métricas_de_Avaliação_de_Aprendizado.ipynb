```python
# Projeto: Cálculo de Métricas de Avaliação de Aprendizado

# 1. Definição arbitrária da Matriz de Confusão
# VP: Verdadeiros Positivos, VN: Verdadeiros Negativos
# FP: Falsos Positivos, FN: Falsos Negativos
vp = 50
vn = 40
fp = 5
fn = 5

# N: Total de elementos
n = vp + vn + fp + fn

# 2. Implementação das Fórmulas de Avaliação
# Sensibilidade (Recall): VP / (VP + FN)
sensibilidade = vp / (vp + fn)

# Especificidade: VN / (FP + VN)
especificidade = vn / (fp + vn)

# Acurácia: (VP + VN) / N
acuracia = (vp + vn) / n

# Precisão (P): VP / (VP + FP)
precisao = vp / (vp + fp)

# F-score: 2 * (P * S) / (P + S)
f_score = 2 * (precisao * sensibilidade) / (precisao + sensibilidade)

# 3. Exibição dos Resultados
print(f"{'Métrica':<20} | {'Valor':<10}")
print("-" * 35)
print(f"{'Sensibilidade':<20} | {sensibilidade:.4f}")
print(f"{'Especificidade':<20} | {especificidade:.4f}")
print(f"{'Acurácia':<20} | {acuracia:.4f}")
print(f"{'Precisão':<20} | {precisao:.4f}")
print(f"{'F-score':<20} | {f_score:.4f}")
```
